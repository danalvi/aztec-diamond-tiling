```python
import math
import pickle
import numpy as np
from itertools import product
```

```python
def G(x,y,L) :
    def e_(n, t) :
        return 2 * np.prod([math.sin(math.pi * n[i] * t[i] ) for i in [0,1]])
    
    def e_L(n, k) :
        k = tuple(k_ / L for k_ in k)
        return e_(n,k)

    def lambda_(n, L) :
        return -2 * sum([(math.sin( (math.pi * n[i]) / (2 * L) ))**2 for i in [0, 1]] )
    
    return - sum([(((lambda_(n, L))**(-1))*e_L(n,x)*e_L(n,y)) for n in product(range(1,L), range(1,L))])
```

```python
fileObj = open('data_50.obj', 'rb')
height_functions = pickle.load(fileObj)
fileObj.close()
```

```python
import time
start_time = time.time()

A = []

for size in sizes :
    A_ = np.ones((2*DIAMOND_SIZE,2*DIAMOND_SIZE))
    for i in range(-DIAMOND_SIZE,DIAMOND_SIZE) :
        for j in range(-DIAMOND_SIZE, DIAMOND_SIZE) :
            if abs(i) + abs(j) <= DIAMOND_SIZE :
                heights_fixed_ij = [height_functions[size][k][(i,j)] for k in range(size)]
                A_[i+ DIAMOND_SIZE,j+DIAMOND_SIZE] = np.var(heights_fixed_ij)
            else :
                A_[i+ DIAMOND_SIZE,j+ DIAMOND_SIZE] = 0
    A.append(A_)
                
print("--- %s minutes ---" % str((time.time() - start_time)/60) )
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2,5,figsize=(16,6))
count = 0
for axrow in axs:
    for ax_ in axrow:
        if (count < 11) :
            pos = ax_.imshow(A[count], cmap='hot', interpolation='nearest')
            fig.colorbar(pos, ax=ax_)
            ax_.title.set_text("# = {}".format(sizes[count]))
        count += 1
```

```python
L=10

A = np.ones((2*L,2*L))
for i in range(-L, L) :
    for j in range(-L, L) :
        A[i,j] = (1/(L**2))*G((i/L,j/L),(i/L,j/L), L)
```
